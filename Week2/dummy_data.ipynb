# ダミーデータの作成
## 【問題1】乱数の作成
---
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

\begin{eqnarray}
\left[
\begin{array}{ccc}
1.0 & 0.8 \\
0.8 & 1.0 \\
\end{array}
\right]
\end{eqnarray}

```python

import numpy as np

# 平均
mean = np.array([-3, 0])
# 共分散
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

# 2次元正規分布による乱数を500個生成する
np.random.seed(0)
data = np.random.multivariate_normal(mean, cov, 500)

print(data.shape)
print(np.cov(data, rowvar=False))

```

## 【問題2】散布図による可視化
---
問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt
%matplotlib inline

x, y = data.T

plt.xlabel("x1", fontsize=15)
plt.ylabel("x2", fontsize=15)
plt.title("scatter", fontsize=15)
plt.tick_params(labelsize=15)
plt.scatter(x, y)
plt.show()
```

## 【問題3】ヒストグラムによる可視化
---
問題1で作成したデータをヒストグラムにより可視化してください。

ヒストグラムはplt.hist()を使うことで描けます。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 平均
mean = np.array([-3, 0])
# 共分散
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

# 2次元正規分布による乱数を500個生成する
np.random.seed(0)
data = np.random.multivariate_normal(mean, cov, 500)

# ヒストグラムを表示

plt.xlabel("x1, x2", fontsize=15)
plt.ylabel("frequency", fontsize=15)
plt.title("histogram of x1, x2", fontsize=15)
plt.tick_params(labelsize=15)
plt.hist(data, bins=50, label=["x1", "x2"])
plt.legend()
plt.grid(linestyle='dotted')
plt.show()
```

### 各次元ごとのヒストグラムを表示

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 平均
mean = np.array([-3, 0])
# 共分散
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

# 2次元正規分布による乱数を500個生成する
np.random.seed(0)
x1, x2 = np.random.multivariate_normal(mean, cov, 500).T

# 2つのヒストグラムを表示
# x1
plt.xlabel("x1", fontsize=15)
plt.ylabel("frequency", fontsize=15)
plt.title("histogram of x1", fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-6, 3) # x軸の表示範囲
plt.hist(x1, bins=50 )
plt.grid(linestyle='dotted')
plt.show()

# x2
plt.xlabel("x2", fontsize=15)
plt.ylabel("frequency", fontsize=15)
plt.title("histogram of x2", fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-6, 3) # x軸の表示範囲
plt.hist(x2, bins=50 )
plt.grid(linestyle='dotted')
plt.show()
```

## 問題4】データの追加
---
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

\begin{eqnarray}
\left[
\begin{array}{ccc}
1.0 & 0.8 \\
0.8 & 1.0 \\
\end{array}
\right]
\end{eqnarray}

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは<font color="red">0</font>、問題4のものは<font color="red">1</font>を表示してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 平均
mean1 = np.array([-3, 0])
mean2 = np.array([0, -3])
# 共分散
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

# それぞれの2次元正規分布による乱数を500個生成する
np.random.seed(0)
x1, y1 = np.random.multivariate_normal(mean1, cov, 500).T
x2, y2 = np.random.multivariate_normal(mean2, cov, 500).T

plt.xlabel("x1", fontsize=15)
plt.ylabel("x2", fontsize=15)
plt.title("scatter", fontsize=15)
plt.tick_params(labelsize=15)
plt.scatter(x1, y1, label="0")
plt.scatter(x2, y2, label="1")
plt.legend()
plt.show()
```

## 【問題5】データの結合
---
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
import numpy as np

# 平均
mean1 = np.array([-3, 0])
mean2 = np.array([0, -3])
# 共分散
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

# それぞれの2次元正規分布による乱数を500個生成する
np.random.seed(0)
data1 = np.random.multivariate_normal(mean1, cov, 500)
data2 = np.random.multivariate_normal(mean2, cov, 500)

# 2つのデータを結合
data3 = np.concatenate([data1, data2])
print("問題1と問題4のデータを結合した後の形状(ndarray)：{}".format(data3.shape))
```

## 【問題6】ラベル付け
---
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。

0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

```python
# 各データにラベル用の列を追加
data1_labeled = np.insert(data1, 2, 0, axis=1)
data2_labeled = np.insert(data2, 2, 1, axis=1)

# 2つのデータを結合
data3_labeled = np.concatenate([data1_labeled, data2_labeled])

print("問題1と問題4のラベル付きのデータを結合した後の形状(ndarray)：{}".format(data3_labeled.shape))
```
